```python
# Convert to cartesian coordinates
radial = np.sqrt(aa**2 + bb**2)
angle = np.arctan2(bb,aa)
```

```python
# Visualize the radial coordinates
plt.pcolormesh(aa,bb,radial)
```

```python
# Visualize the angle coordinates
plt.pcolormesh(bb,aa,angle)
```

```python
#Calculate the quantity
d = np.cos(4*radial)**2
e = np.sin(4*angle)**2
f = e + d
f
```

```python
# Plot f on the x/y plane
plt.pcolormesh(aa, bb, f)
```

```python
# finding the means of all y values
f_ymean = f.mean(axis=1) 
f_ymean
```

```python
# finding the mean of all x values
f_xmean= f.mean(axis=0)
f_xmean
```

### 2.3 Make a plot for each column of data in T, S and P (three plots).

```python
plt.figure(figsize=(10, 10))
plt.grid(True, c = 'grey')
depNeg = dep*-1
plt.plot(t, depNeg)
plt.xlabel('Temperature')
plt.ylabel('Depth')
plt.title('Temperature vs Depth')
plt.xticks(range(3, 26, 2))
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
```

```python
plt.figure(figsize=(10, 10))
plt.grid(True, c = 'grey')
plt.plot(s, depNeg)
plt.xlabel('Salinity')
plt.ylabel('Depth')
plt.title('Salinity vs Depth')
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
```

```python
plt.figure(figsize=(10, 10))
plt.grid(True, c = 'grey')
plt.plot(p, depNeg)
plt.xlabel('Pressure')
plt.ylabel('Depth')
plt.title('Pressure vs Depth')
```

### 2.4, 2.5 Compute the mean and standard deviation of each of T, S and P at each depth in level. M make three similar plsot, but show only the mean T, S and P at each depth. Show error bars on each plot using the standard deviation')

```python
plt.figure(figsize=(10, 10))
plt.grid(True, c = 'grey')
tMean = t.mean(axis=1)
plt.plot(tMean, depNeg)
tStd = t.std(axis=1)
plt.errorbar(tMean, depNeg, yerr=tStd, fmt='-o', capsize=4)
plt.title('Mean Temperature vs Depth')
plt.xlabel('Mean Temperature')
plt.ylabel('Depth')
plt.title('Mean Temp vs Depth')
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
```

```python
plt.figure(figsize=(10, 10))
sMean = s.mean(axis=1)
plt.plot(sMean, depNeg)
sStd = s.std(axis=1)
plt.errorbar(sMean, depNeg, yerr=sStd, fmt='s', capsize=6, c = '#E81828')
plt.grid(True, c = 'grey')
plt.title('Mean Salinity vs Depth')
plt.xlabel('Mean Salinity')
plt.ylabel('Depth')
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
```

```python
plt.figure(figsize=(10, 10))
pMean = p.mean(axis=1)
plt.plot(p, depNeg)
pStd = p.std(axis=1)
plt.errorbar(pMean, depNeg, yerr=pStd, fmt='*', capsize=5, c = '#E81828')
plt.grid(True, c = 'grey')
plt.title('Mean Pressure vs Depth')
plt.xlabel('Mean Pressure')
plt.ylabel('Depth')
```

### 2.6 Account For Missing Data

```python
plt.figure(figsize=(10, 10))
tMean = np.nanmean(t, axis=1) 
plt.plot(tMean, depNeg)
tStd = np.nanstd(t, axis=1)
plt.errorbar(tMean, depNeg, yerr=tStd, fmt='-o', capsize=6, c = '#E81828')
plt.grid(True, c = 'grey')
plt.xlabel('Mean Temperature')
plt.ylabel('Depth')
plt.title('Mean Temperature vs Depth (II)')
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
```

```python
plt.figure(figsize=(10, 10))
sMean = np.nanmean(s, axis=1)
plt.plot(sMean, depNeg)
sStd = np.nanstd(s, axis=1)
plt.errorbar(sMean, depNeg, yerr=sStd, fmt='d', capsize=6, c = '#E81828')
plt.grid(True, c = 'grey')
plt.xlabel('Mean Salinity')
plt.ylabel('Depth')
plt.title('Mean Salinity vs Depth (II)')
plt.gca().xaxis.set_ticks_position('top')
plt.gca().xaxis.set_label_position('top')
```

```python
plt.figure(figsize=(10, 10))
pMean = np.nanmean(p, axis=1) 
plt.plot(pMean,depNeg)
pStd = np.nanstd(p, axis=1)
plt.errorbar(pMean, depNeg, yerr=pStd, fmt='d', capsize=4, c = '#E81828')
plt.grid(True, c = 'grey')
plt.xlabel('Mean Pressure')
plt.ylabel('Depth')
plt.title('Mean Pressure vs Depth (II)')
```
